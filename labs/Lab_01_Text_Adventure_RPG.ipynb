````python
import random

def roll_d20():
    """Roll a 20-sided die."""
    return random.randint(1, 20)

def roll_dice(num_dice, sides):
    """Roll multiple dice and return the total. E.g., roll_dice(2, 6) for 2d6."""
    return sum(random.randint(1, sides) for _ in range(num_dice))


class Character:
    """Base class for all characters in the game."""
    
    def __init__(self, name, health, strength, defense):
        self.name = name
        self.max_health = health
        self.health = health
        self.strength = strength
        self.defense = defense
        self.infection_level = 0  # 0-100 scale, where 100 is fully infected
    
    def is_alive(self):
        """Return True if the character is still alive."""
        return self.health > 0
    
    def take_damage(self, amount):
        """Reduce health by damage amount, but don't go below 0."""
        self.health = max(0, self.health - amount)
        print(f"{self.name} takes {amount} damage! (HP: {self.health}/{self.max_health})")
        
        if self.infection_level > 0:
            status = "🦠" if self.infection_level >= 50 else "⚠️"
            print(f"{status} Infection level: {self.infection_level}%")
        
        if not self.is_alive():
            print(f"{self.name} has fallen!")
    
    def apply_infection(self, amount):
        """Apply infection damage. At 100%, character is incapacitated."""
        self.infection_level = min(100, self.infection_level + amount)
        if amount > 0:
            print(f"🦠 {self.name} becomes infected! (Infection: {self.infection_level}%)")
        if self.infection_level >= 100:
            print(f"💀 {self.name} has succumbed to infection!")
            self.health = 0
    
    def attack(self, target):
        """Perform a D20 attack against a target."""
        print(f"\n{self.name} attacks {target.name}!")
        
        attack_roll = roll_d20()
        total_attack = attack_roll + self.strength
        
        print(f"Roll: {attack_roll} + STR({self.strength}) = {total_attack}")
        print(f"{target.name}'s Defense: {target.defense}")
        
        if total_attack > target.defense:
            print("Hit!")
            
            # Option 1: Fixed damage equal to strength
            damage = self.strength
            
            # Optional: Replace with dice-based damage
            # damage = roll_dice(1, 6) + self.strength
            
            target.take_damage(damage)
        else:
            print("Miss!")
    
    def __str__(self):
        return f"{self.name} (HP: {self.health}/{self.max_health}, STR: {self.strength}, DEF: {self.defense})"
````

### 2.2 Player and Enemy Subclasses

Now create subclasses that inherit from `Character`:

**Player class** might have:
- An `inventory` list
- Methods: `pick_up(item)`, `use_item(item_name)`, `show_inventory()`
- A special ability (heal, power attack, etc.)

**Enemy class** might have:
- An `xp_value` (experience points when defeated)
- A `loot` attribute (what they drop)
- Override `attack()` with different behavior

**Specific enemy types** (Minion, Elite, Boss) can have:
- Different base stats
- Unique abilities
- Different attack messages (make it fun!)

````python
# PLAYER CLASS

class Player(Character):
    """The player character."""
    
    def __init__(self, name):
        # Starting stats tuned for zombie survival
        super().__init__(name, health=100, strength=8, defense=12)
        self.inventory = []
        self.xp = 0
        self.crit_chance = 0.20  # 20% chance for critical hit
        self.equipped_weapon = None  # No weapon equipped initially (fighting with fists)
    
    def pick_up(self, item):
        """Add an item to inventory."""
        self.inventory.append(item)
        
        # Show a clear, celebratory message for picking up items
        weapon_indicator = " ⚔️ [WEAPON]" if item in WEAPONS else ""
        print(f"\n✅ You picked up: {item}{weapon_indicator}")
        
        # Give contextual hints
        if item in WEAPONS:
            print(f"💡 Tip: Type 'equip {item.lower()}' to use this weapon in combat")
    
    def show_inventory(self):
        """Display inventory contents."""
        print(f"\n{'='*50}")
        print(f"📦 {self.name}'s Inventory")
        print(f"{'='*50}")
        
        if self.equipped_weapon:
            print(f"⚔️  Equipped: {self.equipped_weapon}")
        else:
            print(f"⚔️  Equipped: Fists (Damage: {self.strength})")
        
        print()
        if not self.inventory:
            print("Inventory is empty.")
        else:
            print("Items:")
            for item in self.inventory:
                weapon_marker = " [WEAPON]" if item in WEAPONS else ""
                print(f"  - {item}{weapon_marker}")
        print(f"{'='*50}")
    
    def equip_weapon(self, weapon_name):
        """Equip a weapon from inventory."""
        if weapon_name not in self.inventory:
            print(f"You don't have '{weapon_name}' in your inventory.")
            return
        
        if weapon_name not in WEAPONS:
            print(f"'{weapon_name}' is not a weapon.")
            return
        
        self.equipped_weapon = WEAPONS[weapon_name]
        print(f"✓ {self.name} equipped {self.equipped_weapon}")
    
    def use_item(self, item_name):
        """Use an item if it exists."""
        if item_name in self.inventory:
            if item_name.lower() == "medkit":
                heal_amount = 20
                self.health = min(self.max_health, self.health + heal_amount)
                print(f"{self.name} used a Medkit and healed {heal_amount} HP!")
                self.inventory.remove(item_name)
            else:
                print(f"{item_name} cannot be used right now.")
        else:
            print(f"{item_name} not found in inventory.")
    
    def attack(self, target):
        """Perform a D20 attack with critical hit chance and weapon damage."""
        print(f"\n{self.name} attacks {target.name}!")
        
        attack_roll = roll_d20()
        total_attack = attack_roll + self.strength
        
        print(f"Roll: {attack_roll} + STR({self.strength}) = {total_attack}")
        print(f"{target.name}'s Defense: {target.defense}")
        
        # Check for critical hit
        is_critical = random.random() < self.crit_chance
        
        if total_attack > target.defense:
            # Base damage is strength
            base_damage = self.strength
            
            # Add weapon damage if equipped
            weapon_damage = 0
            if self.equipped_weapon:
                weapon_damage = self.equipped_weapon.get_damage()
                print(f"💥 Weapon bonus: +{weapon_damage} from {self.equipped_weapon.name}!")
            
            total_damage = base_damage + weapon_damage
            
            # Apply critical hit multiplier
            if is_critical:
                total_damage = total_damage * 2  # Double damage on crit
                print(f"⚡ CRITICAL HIT! {self.name} strikes with devastating force!")
            else:
                print("Hit!")
            
            target.take_damage(total_damage)
        else:
            print("Miss!")
    
    def power_attack(self, target):
        """Special ability: stronger but riskier attack."""
        print(f"\n{self.name} uses POWER ATTACK on {target.name}!")
        
        attack_roll = roll_d20()
        total_attack = attack_roll + self.strength + 3  # bonus damage
        
        print(f"Roll: {attack_roll} + STR({self.strength}) + 3 = {total_attack}")
        
        if total_attack > target.defense:
            damage = self.strength + roll_dice(1, 6)
            print("Critical Strike!")
            target.take_damage(damage)
        else:
            print("Power attack missed!")


# -------------------------------------
# ENEMY BASE CLASS
# -------------------------------------

class Enemy(Character):
    """Base class for enemies."""
    
    def __init__(self, name, health, strength, defense, xp_value=10, loot=None, infection_chance=0.15):
        super().__init__(name, health, strength, defense)
        self.xp_value = xp_value
        self.loot = loot if loot else []
        self.infection_chance = infection_chance  # Chance to infect on hit
    
    def attack(self, target):
        """Enemies attack with flavor text and infection chance."""
        print(f"\n{self.name} lunges at {target.name}!")
        
        attack_roll = roll_d20()
        total_attack = attack_roll + self.strength
        
        if total_attack > target.defense:
            damage = roll_dice(1, 4) + self.strength
            print(f"{self.name} claws viciously!")
            target.take_damage(damage)
            
            # Check for infection
            if random.random() < self.infection_chance:
                infection_amount = roll_dice(1, 3) + 5  # 6-8% infection per hit
                target.apply_infection(infection_amount)
        else:
            print(f"{self.name}'s attack misses!")


# -------------------------------------
# SPECIFIC ENEMY TYPES
# -------------------------------------

class Minion(Enemy):
    """Weak zombie."""
    
    def __init__(self):
        super().__init__(
            name="Walker",
            health=30,
            strength=4,
            defense=8,
            xp_value=15,
            loot=["Rotten Food"],
            infection_chance=0.20  # 20% chance to infect
        )


class Elite(Enemy):
    """Tougher, faster enemy."""
    
    def __init__(self):
        super().__init__(
            name="Runner",
            health=50,
            strength=7,
            defense=11,
            xp_value=30,
            loot=["Medkit", "Ammo"],
            infection_chance=0.25  # 25% chance to infect
        )


class Boss(Enemy):
    """Final boss enemy."""
    
    def __init__(self):
        super().__init__(
            name="Mutated Alpha",
            health=120,
            strength=12,
            defense=14,
            xp_value=100,
            loot=["Radio Part", "Rare Weapon"],
            infection_chance=0.35  # 35% chance to infect
        )
````

### 3.1 The Location Class

Each location needs:
- **Attributes**: `name`, `description`, `connections` (dict mapping directions to other locations), `enemies` (list), `items` (list)
- **Methods**:
  - `describe()`: Print the location description and what's here
  - `get_exits()`: Return available directions
  - `add_connection(direction, location)`: Connect to another location
  
**Pro tip**: Store connections as a dictionary: `{"north": cave, "south": village}`

````python
# YOUR CODE: Build the Location class

class Location:
    """A location in the game world."""
    
    def __init__(self, name, description):
        self.name = name
        self.description = description
        self.connections = {}  # {"north": Location, "south": Location, etc.}
        self.enemies = []      # List of enemies in this location
        self.items = []        # List of items in this location
    
    def describe(self):
        """Print a full description of the location."""
        print(f"\n{'='*50}")
        print(f"📍 {self.name}")
        print(f"{'='*50}")
        print(self.description)
        
        # Show enemies if present
        if self.enemies:
            print("\n⚠️ ENEMIES HERE:")
            for enemy in self.enemies:
                print(f"  - {enemy.name} (HP: {enemy.health}/{enemy.max_health})")
        
        # Show items if present
        if self.items:
            print("\n💰 ITEMS YOU CAN PICK UP:")
            for item in self.items:
                weapon_indicator = " [WEAPON]" if item in WEAPONS else ""
                print(f"  - {item}{weapon_indicator}")
        
        # Show available exits prominently
        print(f"\n🧭 DIRECTIONS YOU CAN TRAVEL:")
        if self.connections:
            directions = list(self.connections.keys())
            for direction in directions:
                destination = self.connections[direction]
                print(f"  → {direction.upper()} - leads to {destination.name}")
        else:
            print("  (No exits available)")
        
        # End with clear call to action
        print("\n" + "="*50)
        print("❓ WHAT DO YOU DO?")
        print("="*50)
        
        # Show context-specific actions
        if self.items:
            print(f"📦 Pick up items:")
            for item in self.items[:2]:  # Show first 2 items as examples
                print(f"   → take {item.lower()}")
        
        if self.connections:
            directions_list = ", ".join(list(self.connections.keys())[:3])
            print(f"🗺️  Explore: {directions_list}")
        
        if self.enemies:
            print(f"⚔️  Combat: fight")
        
        print(f"📋 Other: inventory, look, help, quit")
        print("="*50)
    
    def get_exits(self):
        """Return a list of available directions."""
        return list(self.connections.keys())
    
    def add_connection(self, direction, location):
        """Connect this location to another."""
        self.connections[direction] = location
````

## Part 4: The Combat System

Time to make battles exciting! We'll implement a turn-based combat system using a simple state machine.

### 4.1 Combat States

Combat flows through states. For example, here is one simple way to do it:
1. **PLAYER_TURN**: Player chooses an action
2. **ENEMY_TURN**: Enemy attacks
3. **COMBAT_END**: Someone won (or ran away)

```
┌─────────────────────────────────────────────────────────┐
│                                                         │
│  [PLAYER_TURN] ─── attack ──→ [ENEMY_TURN]              │
│       │                            │                    │
│       │                            │                    │
│       │←─────── (enemy alive) ─────┘                    │
│       │                            │                    │
│       │                            ▼                    │
│       │                      (enemy dead?)              │
│       │                            │                    │
│       └── run away ─────→ [COMBAT_END] ←────────────────┘
│                                                         │
└─────────────────────────────────────────────────────────┘
```

### 4.1 Enhanced Combat System Features

The improved combat system now includes realistic battle mechanics:

**Player Actions (choose ONE per turn):**
- **(A)ttack**: Attack the enemy with hit → crit → damage resolution
  - Roll d20 + STR vs enemy DEF
  - Check for critical hit (20% chance = 2x damage)
  - Apply damage + status effects
  
- **(D)efend**: Take a defensive stance
  - Reduces incoming damage by 50% this turn
  - Guard is temporary (resets each turn)
  - Good for when you're low on health
  
- **(U)se**: Use an item from your inventory
  - Medkit = heal 20 HP
  - Can be used in combat!
  - Items consumed after use
  
- **(R)un**: Attempt to flee from combat
  - Base 50% success chance
  - Infection level > 50% reduces escape chance by 15%
  - Failure means enemy gets a free attack

**Combat Flow:**
1. PLAYER_TURN → Player chooses action
2. RESOLVE → Damage calculated, critical hits checked, status applied
3. END_OF_TURN_EFFECTS → Guard expires, DOT ticks, effects decay
4. ENEMY_TURN → Enemy attacks (with infection chance)
5. CHECK_END → Someone dead? Continue or end combat

**Realistic Mechanics:**
- ✓ Status effects (infection) affect gameplay (slows escape)
- ✓ Guard/defend provides tactical depth
- ✓ Items give players resources to manage
- ✓ Infection is tracked throughout combat
- ✓ Critical hits reward good rolls

```

│                     [COMBAT_END]  ───→ [PLAYER_TURN]                │
│                                                                    │
└────────────────────────────────────────────────────────────────────┘
```

````python
# YOUR CODE: Implement the combat system

class Combat:
    """Manages turn-based combat between player and enemy."""
    
    # Combat states
    PLAYER_TURN = "player_turn"
    ENEMY_TURN = "enemy_turn"
    COMBAT_END = "combat_end"
    
    def __init__(self, player, enemy):
        self.player = player
        self.enemy = enemy
        self.state = Combat.PLAYER_TURN
        self.combat_log = []
        
        # Track temporary combat states
        self.player_guarding = False
        self.enemy_guarding = False
    
    def start(self):
        """Begin combat and run until someone wins/loses/flees."""
        print(f"\n⚔️ COMBAT BEGINS! ⚔️")
        print(f"{self.player.name} vs {self.enemy.name}!")
        print("="*60)
        
        while self.state != Combat.COMBAT_END:
            if self.state == Combat.PLAYER_TURN:
                self.player_turn()
            elif self.state == Combat.ENEMY_TURN:
                self.enemy_turn()
        
        return self.get_result()
    
    def player_turn(self):
        """Handle player's turn in combat with multiple action options."""
        print(f"\n{self.player} | {self.enemy}")
        print("\nWhat do you do?")
        print("  (a)ttack  - Deal damage to the enemy")
        print("  (d)efend  - Reduce incoming damage this turn")
        print("  (u)se     - Use an item from inventory")
        print("  (r)un     - Attempt to flee combat")
        
        action = input("> ").lower().strip()
        
        if action in ["a", "attack"]:
            self.resolve_player_attack()
            
        elif action in ["d", "defend"]:
            self.player_guarding = True
            print(f"\n🛡️  {self.player.name} takes a defensive stance!")
            print("Incoming damage will be reduced this turn.")
            self.state = Combat.ENEMY_TURN
            
        elif action in ["u", "use"]:
            self.player_use_item()
            # After using item, still proceed to enemy turn
            if self.enemy.is_alive():
                self.state = Combat.ENEMY_TURN
            
        elif action in ["r", "run"]:
            self.attempt_escape()
            
        else:
            print("Invalid action. Try (a)ttack, (d)efend, (u)se, or (r)un.")
    
    def resolve_player_attack(self):
        """Resolve player attack with hit check, crit check, and damage."""
        print(f"\n{self.player.name} attacks {self.enemy.name}!")
        
        # 1. ATTACK ROLL
        attack_roll = roll_d20()
        total_attack = attack_roll + self.player.strength
        
        print(f"Roll: {attack_roll} + STR({self.player.strength}) = {total_attack}")
        print(f"{self.enemy.name}'s Defense: {self.enemy.defense}")
        
        # 2. HIT CHECK
        if total_attack > self.enemy.defense:
            # 3. CRITICAL HIT CHECK
            is_critical = random.random() < self.player.crit_chance
            
            # 4. CALCULATE DAMAGE
            if is_critical:
                damage = self.player.strength * 2
                print(f"⚡ CRITICAL HIT! Damage: {damage}")
            else:
                damage = self.player.strength
                print(f"Hit! Damage: {damage}")
            
            # 5. APPLY DAMAGE & ON-HIT EFFECTS
            self.enemy.take_damage(damage)
            
            # Check for infection chance (zombie attack mechanic reversed)
            # Instead, could add other status effects here if desired
            
        else:
            print("Miss! No damage dealt.")
        
        # Attack resolves, move to enemy turn if enemy still alive
        if self.enemy.is_alive():
            self.state = Combat.ENEMY_TURN
        else:
            print(f"\n🎉 {self.enemy.name} has been defeated!")
            self.state = Combat.COMBAT_END
    
    def player_use_item(self):
        """Allow player to use items during combat."""
        if not self.player.inventory:
            print("Your inventory is empty!")
            return
        
        print("\nWhat item do you want to use?")
        for i, item in enumerate(self.player.inventory, 1):
            print(f"  {i}. {item}")
        print(f"  {len(self.player.inventory) + 1}. Cancel")
        
        try:
            choice = int(input("> "))
            if choice == len(self.player.inventory) + 1 or choice < 1:
                print("You decide not to use anything.")
                return
            
            item = self.player.inventory[choice - 1]
            
            # Apply item effects
            if item.lower() == "medkit":
                heal_amount = 20
                old_hp = self.player.health
                self.player.health = min(self.player.max_health, self.player.health + heal_amount)
                actual_heal = self.player.health - old_hp
                print(f"✨ {self.player.name} used a Medkit and healed {actual_heal} HP!")
                self.player.inventory.pop(choice - 1)
            else:
                print(f"You can't use {item} right now in combat.")
                return
        
        except (ValueError, IndexError):
            print("Invalid choice.")
    
    def attempt_escape(self):
        """Attempt to flee combat. Success depends on speed/luck."""
        print(f"\n{self.player.name} attempts to flee!")
        
        # 50% base chance, can be modified by infection level (slower when infected)
        escape_chance = 0.5
        if self.player.infection_level > 50:
            escape_chance -= 0.15  # Harder to escape when heavily infected
            print("🦠 Your infection slows you down...")
        
        if random.random() < escape_chance:
            print("✅ You successfully fled combat!")
            self.state = Combat.COMBAT_END
        else:
            print("❌ You couldn't get away! The enemy blocks your path!")
            print(f"{self.enemy.name} takes this as an opening...")
            self.state = Combat.ENEMY_TURN
    
    def enemy_turn(self):
        """Handle enemy's turn in combat."""
        print(f"\n{self.enemy.name}'s turn...")
        
        # Simple AI: enemy can attack or guard
        # For now, just always attack
        self.resolve_enemy_attack()
        
        # 5. END OF TURN EFFECTS (reduce guard, apply DOT, etc.)
        self.apply_end_of_turn_effects()
        
        # Check if player is still alive
        if not self.player.is_alive():
            print(f"\n💀 {self.player.name} has fallen!")
            self.state = Combat.COMBAT_END
        else:
            self.state = Combat.PLAYER_TURN
    
    def resolve_enemy_attack(self):
        """Resolve enemy attack with damage reduction for guarding."""
        print(f"\n{self.enemy.name} attacks {self.player.name}!")
        
        # 1. ATTACK ROLL
        attack_roll = roll_d20()
        total_attack = attack_roll + self.enemy.strength
        
        if total_attack > self.player.defense:
            # 2. CALCULATE BASE DAMAGE
            base_damage = roll_dice(1, 4) + self.enemy.strength
            
            # 3. APPLY GUARD REDUCTION IF ACTIVE
            if self.player_guarding:
                damage = max(1, base_damage // 2)  # Half damage when guarding
                print(f"{self.enemy.name} claws viciously!")
                print(f"🛡️  You block most of the damage! ({base_damage} → {damage})")
            else:
                damage = base_damage
                print(f"{self.enemy.name} claws viciously!")
            
            # 4. APPLY DAMAGE & ON-HIT EFFECTS
            self.player.take_damage(damage)
            
            # 5. CHECK FOR INFECTION
            if random.random() < self.enemy.infection_chance:
                infection_amount = roll_dice(1, 3) + 5
                self.player.apply_infection(infection_amount)
        else:
            print(f"{self.enemy.name}'s attack misses!")
    
    def apply_end_of_turn_effects(self):
        """Apply end-of-turn effects: clear guard, apply status effects, regen, etc."""
        # Clear guard status for next turn
        if self.player_guarding:
            self.player_guarding = False
            # Guard is one-turn only
        
        if self.enemy_guarding:
            self.enemy_guarding = False
        
        # Could add other effects here:
        # - Poison/bleed damage over time
        # - Regeneration
        # - Status effect duration decay
        # - Etc.
    
    def get_result(self):
        """Return the combat result: 'victory', 'defeat', or 'fled'."""
        if not self.enemy.is_alive():
            return "victory"
        elif not self.player.is_alive():
            return "defeat"
        else:
            return "fled"
````

## Part 5: The Main Game Loop

Now let's tie it all together! The `Game` class manages the overall game state and the main loop.

### 5.1 Game States

Your main game should have these states:
- **EXPLORING**: Player is moving around, examining locations
- **IN_COMBAT**: Player is fighting an enemy
- **GAME_OVER**: Player died
- **VICTORY**: Player won!

### 5.2 Available Actions

When exploring, the player might:
- Move: `go north`, `go south`, `go east`, `go west`
- Look around: `look`
- Fight: `fight` or `attack`
- Inventory: `inventory` or `i`
- Get help: `help`
- Quit: `quit`

Handle invalid input gracefully!

````python
# YOUR CODE: Build the main Game class

class Game:
    """Main game controller with state management."""
    
    # Game states
    EXPLORING = "exploring"
    IN_COMBAT = "in_combat"
    GAME_OVER = "game_over"
    VICTORY = "victory"
    
    def __init__(self):
        self.player = None
        self.current_location = None
        self.state = Game.EXPLORING  # Start in exploring state
        self.game_running = True
    
    def start(self):
        """Initialize and start the game."""
        self.show_intro()
        self.create_player()
        self.current_location = create_world()  # Load world
        self.current_location.describe()
        print("\n💡 QUICK START GUIDE:")
        print("   1. Type 'take rusty bat' to pick up your first weapon")
        print("   2. Type 'equip rusty bat' to equip it")
        print("   3. Type 'north' to explore Downtown")
        
        # Main game loop - continues until game ends
        while self.game_running:
            if self.state == Game.EXPLORING:
                self.exploration_loop()
            elif self.state == Game.IN_COMBAT:
                # Combat state is handled through initiate_combat transition
                pass
            elif self.state == Game.GAME_OVER:
                self.show_game_over()
                break
            elif self.state == Game.VICTORY:
                self.show_victory()
                break
    
    def show_intro(self):
        """Display the game introduction."""
        print("\n" + "="*60)
        print("    🧟 ZOMBIE SURVIVAL: TEXT ADVENTURE 🧟")
        print("="*60)
        print("\nYou wake up in a barricaded safehouse.")
        print("The city outside is overrun with undead.")
        print("Your goal: Reach the radio tower, get the parts, and escape!")
        print("\n" + "="*60)
    
    def create_player(self):
        """Create the player character."""
        print("\nWhat is your name, survivor?")
        name = input("> ").strip()
        if not name:
            name = "Survivor"
        self.player = Player(name)
        print(f"\nWelcome, {name}! Your adventure begins...")
        print(f"HP: {self.player.health} | STR: {self.player.strength} | DEF: {self.player.defense}")
    
    def exploration_loop(self):
        """Handle player input during exploration state."""
        self.show_actions()
        command = input("> ").lower().strip()
        
        # Parse the command
        parts = command.split()
        if not parts:
            return
        
        action = parts[0]
        
        if action == "help":
            self.show_help()
        
        elif action == "look":
            self.current_location.describe()
        
        elif action == "go" and len(parts) > 1:
            direction = parts[1]
            self.move(direction)
        
        elif action in ["north", "south", "east", "west", "up", "down"]:
            self.move(action)
        
        elif action in ["fight", "attack"]:
            self.initiate_combat()
        
        elif action in ["inventory", "i"]:
            self.player.show_inventory()
        
        # Handle multiple pickup command variations
        elif action in ["take", "grab"]:
            item_name = " ".join(parts[1:]) if len(parts) > 1 else None
            self.pick_up_item(item_name)
        
        # Handle "pick up [item]" - two word command
        elif action == "pick" and len(parts) > 1 and parts[1] == "up":
            item_name = " ".join(parts[2:]) if len(parts) > 2 else None
            self.pick_up_item(item_name)
        
        # Handle "pickup [item]" - one word command
        elif action == "pickup":
            item_name = " ".join(parts[1:]) if len(parts) > 1 else None
            self.pick_up_item(item_name)
        
        # Handle "get [item]" command
        elif action == "get":
            item_name = " ".join(parts[1:]) if len(parts) > 1 else None
            self.pick_up_item(item_name)
        
        elif action in ["equip", "wield"]:
            weapon_name = " ".join(parts[1:]) if len(parts) > 1 else None
            self.equip_weapon(weapon_name)
        
        elif action == "quit":
            print("Thanks for playing!")
            self.game_running = False
        
        else:
            print("I don't understand that command. Type 'help' for options.")
    
    def show_actions(self):
        """Show available commands every turn with clear examples."""
        print("\n" + "="*60)
        print("⚡ AVAILABLE ACTIONS")
        print("="*60)
        
        # Show context-specific suggestions
        if self.current_location.items:
            print(f"📦 PICK UP ITEMS:")
            print(f"   take [item]          Example: take {self.current_location.items[0].lower()}")
            print(f"   get [item]           Example: get {self.current_location.items[0].lower()}")
            print(f"   pick up [item]       Example: pick up {self.current_location.items[0].lower()}")
        
        print(f"\n🗺️  EXPLORE:")
        available_dirs = list(self.current_location.connections.keys())
        if available_dirs:
            dir_examples = ", ".join(available_dirs[:3])
            print(f"   {dir_examples}           Type direction to move")
            print(f"   go [direction]       Example: go {available_dirs[0]}")
        else:
            print(f"   north, south, east, west")
        
        print(f"\n⚔️  COMBAT:")
        if self.current_location.enemies:
            print(f"   fight                Fight {self.current_location.enemies[0].name} here!")
        else:
            print(f"   fight                Attack an enemy (none here)")
        
        print(f"\n🎒 INVENTORY:")
        print(f"   inventory (or i)     Check your items and equipped weapon")
        if self.player and self.player.inventory:
            weapons = [item for item in self.player.inventory if item in WEAPONS]
            if weapons:
                print(f"   equip [weapon]       Example: equip {weapons[0].lower()}")
        
        print(f"\n❓ OTHER:")
        print(f"   look                 Examine surroundings")
        print(f"   help                 Show detailed help")
        print(f"   quit                 Exit game")
        print("="*60)
    
    def move(self, direction):
        """Move the player in the specified direction (EXPLORING state)."""
        if direction in self.current_location.connections:
            self.current_location = self.current_location.connections[direction]
            print(f"\n🚶 You travel {direction}...")
            self.current_location.describe()
        else:
            print(f"You can't go {direction} from here.")
    
    def pick_up_item(self, item_name=None):
        """Pick up an item from the current location."""
        if not self.current_location.items:
            print("There are no items to pick up here.")
            return
        
        if not item_name:
            print("\n📦 Items available here:")
            for item in self.current_location.items:
                weapon_marker = " [WEAPON]" if item in WEAPONS else ""
                print(f"  - {item}{weapon_marker}")
            print("\nType: take [item name]")
            return
        
        # Case-insensitive matching
        normalized = item_name.lower()
        matched_item = None
        for item in self.current_location.items:
            if item.lower() == normalized:
                matched_item = item
                break
        
        if not matched_item:
            print(f"❌ There is no '{item_name}' here.")
            return
        
        self.current_location.items.remove(matched_item)
        self.player.pick_up(matched_item)
    
    def equip_weapon(self, weapon_name=None):
        """Equip a weapon from the player's inventory."""
        if not weapon_name:
            print("\n🎒 Available weapons in inventory:")
            weapons_found = False
            for item in self.player.inventory:
                if item in WEAPONS:
                    print(f"  - {item} ({WEAPONS[item]})")
                    weapons_found = True
            
            if not weapons_found:
                print("  (No weapons in inventory)")
            
            print("\nType: equip [weapon name]")
            return
        
        # Case-insensitive matching
        normalized = weapon_name.lower()
        matched_weapon = None
        for item in self.player.inventory:
            if item.lower() == normalized:
                matched_weapon = item
                break
        
        if not matched_weapon:
            print(f"❌ You don't have '{weapon_name}' in your inventory.")
            return
        
        self.player.equip_weapon(matched_weapon)
    
    def initiate_combat(self):
        """Start combat with an enemy in the current location."""
        if not self.current_location.enemies:
            print("❌ There's nothing to fight here.")
            return
        
        enemy = self.current_location.enemies[0]  # Fight first enemy
        
        # Before entering combat, give option to run away
        print(f"\n⚠️ A wild {enemy.name} appears!")
        print(f"Enemy HP: {enemy.health}/{enemy.max_health}")
        print(f"Enemy DEF: {enemy.defense}")
        
        if self.player.equipped_weapon:
            print(f"Your weapon: {self.player.equipped_weapon}")
        else:
            print(f"You're fighting with your fists (STR: {self.player.strength})")
        
        print("\n" + "="*60)
        print("⚔️ COMBAT OPTIONS:")
        print("="*60)
        print("1. fight - Enter combat (will take turns fighting)")
        print("2. run   - Attempt to escape (50% success rate)")
        print("="*60)
        
        action = input("\nChoose an action: ").lower().strip()
        
        if action in ["run", "escape", "flee"]:
            # Attempt to run away
            self.attempt_escape(enemy)
        elif action in ["fight", "attack"]:
            # Enter combat
            self.state = Game.IN_COMBAT
            battle = Combat(self.player, enemy)
            result = battle.start()
            
            # Return to exploring state after combat
            self.state = Game.EXPLORING
            
            if result == "victory":
                self.current_location.enemies.remove(enemy)
                self.check_victory_condition(enemy)
            elif result == "defeat":
                self.state = Game.GAME_OVER
        else:
            print("Invalid action. Combat started automatically!")
            self.state = Game.IN_COMBAT
            battle = Combat(self.player, enemy)
            result = battle.start()
            
            self.state = Game.EXPLORING
            
            if result == "victory":
                self.current_location.enemies.remove(enemy)
                self.check_victory_condition(enemy)
            elif result == "defeat":
                self.state = Game.GAME_OVER
    
    def attempt_escape(self, enemy):
        """Attempt to escape from combat with probability."""
        # Base escape chance is 50%
        escape_chance = 50
        
        # Reduce escape chance if heavily infected
        if self.player.infection_level > 50:
            escape_chance -= 15
            print(f"⚠️ Your infection slows you down (-15% escape chance)")
        
        escape_chance = max(10, escape_chance)  # Minimum 10% chance
        
        print(f"\n🏃 Attempting to run away...")
        print(f"Escape chance: {escape_chance}%")
        
        escape_roll = random.randint(1, 100)
        
        if escape_roll <= escape_chance:
            print(f"\n✅ SUCCESS! You escaped from the {enemy.name}!")
            print(f"You run back to the {self.current_location.name}")
        else:
            print(f"\n❌ FAILED! You couldn't escape!")
            print(f"The {enemy.name} blocked your path and forced you into combat!\n")
            
            # Failed escape - locked into combat
            self.state = Game.IN_COMBAT
            battle = Combat(self.player, enemy)
            result = battle.start()
            
            self.state = Game.EXPLORING
            
            if result == "victory":
                self.current_location.enemies.remove(enemy)
                self.check_victory_condition(enemy)
            elif result == "defeat":
                self.state = Game.GAME_OVER
    
    def check_victory_condition(self, enemy):
        """Check if the player meets the win conditions."""
        if not isinstance(enemy, Boss):
            return
        
        at_radio_tower = "Radio Tower" in self.current_location.name
        has_radio_part = "Radio Part" in self.player.inventory
        has_keycard = "Keycard" in self.player.inventory
        
        if at_radio_tower and has_radio_part and has_keycard:
            print("\n📡 You repair the radio and call for extraction!")
            print("🚁 The evac chopper arrives just as the horde closes in...")
            self.state = Game.VICTORY
        else:
            print("\n⚠️ You defeated the boss, but extraction isn't ready yet.")
            if not has_radio_part:
                print("  Missing: Radio Part")
            if not has_keycard:
                print("  Missing: Keycard")
            if not at_radio_tower:
                print("  You must defeat the boss at the Radio Tower")
    
    def show_help(self):
        """Display available commands."""
        print("\n" + "="*70)
        print("📜 AVAILABLE COMMANDS:")
        print("="*70)
        print("  MOVEMENT:")
        print("    go [direction]     - Move in a direction")
        print("    north/south/east/west - Direct movement")
        print("\n  ITEMS:")
        print("    take [item]        - Pick up an item")
        print("    get [item]         - Pick up an item")
        print("    pick up [item]     - Pick up an item")
        print("    grab [item]        - Pick up an item")
        print("\n  INVENTORY & EQUIPMENT:")
        print("    inventory (i)      - Check your inventory")
        print("    equip [weapon]     - Equip a weapon")
        print("\n  COMBAT:")
        print("    fight              - Attack an enemy")
        print("    run                - Attempt to escape (50% success)")
        print("\n  OTHER:")
        print("    look               - Examine your surroundings")
        print("    help               - Show this help message")
        print("    quit               - Exit the game")
        print("="*70)
    
    def show_game_over(self):
        """Display game over message (GAME_OVER state)."""
        print("\n" + "="*60)
        print("💀 GAME OVER - YOU HAVE FALLEN 💀")
        print("="*60)
        print("\nYou have succumbed to the undead infection...")
        print("The zombie apocalypse claims another victim.")
        print("\n(But you can always try again!)")
        print("="*60)
    
    def show_victory(self):
        """Display victory message (VICTORY state)."""
        print("\n" + "="*60)
        print("🎉 VICTORY! YOU HAVE ESCAPED! 🎉")
        print("="*60)
        print(f"\nCongratulations, {self.player.name}!")
        print("You successfully repaired the radio and escaped!")
        print("The helicopter lifted you to safety just in time.")
        print("\nFinal Stats:")
        print(f"  Health: {self.player.health}/{self.player.max_health}")
        print(f"  Infection: {self.player.infection_level}%")
        print(f"  Items Collected: {len(self.player.inventory)}")
        print("="*60)
````
